```python
from IPython.display import clear_output
import numpy as np
import pandas as pd
from ScraperFC.shared_functions import check_season, xpath_soup, sources, UnavailableSeasonException, \
    NoMatchLinksException
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.service import Service as ChromeService
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.action_chains import ActionChains
from selenium.webdriver.support.ui import Select
from urllib.request import urlopen
import requests
import pyautogui
from bs4 import BeautifulSoup
from tqdm import tqdm
import time
import re
from datetime import datetime
import random
```

```python
def iniciar_sesion(url_login, username, password):
    global driver
    global submenu
    options = Options()
    options.headless = True
    options.add_argument(
        'user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) ' + \
        'AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 ' + \
        'Safari/537.36'
    )
    options.add_argument('window-size=1400,600')
    options.add_argument('--incognito')
    prefs = {'profile.managed_default_content_settings.images': 2}  # No carga imagenes
    options.add_experimental_option('prefs', prefs)
    options.add_argument('--log-level=3')
    driver = webdriver.Chrome(
        service=ChromeService(ChromeDriverManager().install()),
        options=options
    )

    # Accedemos a la página de inicio de sesión utilizando Selenium
    driver.get(url_login)

    # espera aleatoria entre 1 y 5 segundos
    time.sleep(random.uniform(1, 5))

    # Ingresamos las credenciales de inicio de sesión
    username_input = driver.find_element(By.NAME, "username")
    password_input = driver.find_element(By.NAME, "password")

    username_input.send_keys(username)
    password_input.send_keys(password)

    # Simulamos el clic en el botón de inicio de sesión
    login_button = driver.find_element(By.CSS_SELECTOR, "#ext-gen40")
    login_button.click()

    # Agregamos las cookies de sesión al controlador de Selenium
    session_cookies = driver.get_cookies()

    # URL de la página que se desea acceder después de iniciar sesión
    url_siguiente = "https://www.vur.gov.co/portal/"

    # Accedemos a la página siguiente utilizando Selenium
    driver.get(url_siguiente)

    # Agregamos las cookies de sesión a la nueva página
    for cookie in session_cookies:
        driver.add_cookie(cookie)

    # Verificamos que se haya accedido a la página correspondiente
    if driver.current_url == "https://www.vur.gov.co/portal/":
        print("Acceso exitoso a la página siguiente.")
    else:
        print("Error al acceder a la página siguiente.")

    # Damos clic en el menu de navegación "Consultas"
    boton_menu = WebDriverWait(driver, 10).until(EC.element_to_be_clickable((By.XPATH, "//*[@id='menu-navegacion']/li[3]/a")))
    boton_menu.click()

    # localiza el submenú que deseas desplegar
    submenu = driver.find_element(By.XPATH, '//*[@id="menu-navegacion"]/li[3]/ul/li/a')
```

```python
def buscar_inmueble(driver, numero_circulo, numero_matricula):
    global select
    # cambiar al contexto del iframe
    iframe = driver.find_element(By.XPATH, '//*[@id="page"]')
    driver.switch_to.frame(iframe)

    # esperar a que aparezca el botón y hacer clic en él
    wait = WebDriverWait(driver, 10)
    boton = wait.until(EC.element_to_be_clickable((By.XPATH, '//*[@id="circulo"]')))
    boton.click()

    # Obtenemos todos los elementos de la lista
    options = driver.find_elements(By.XPATH, '//div[@id="mySelect_chosen"]/div/ul/li')

    # Encontrar el elemento del menú desplegable
    dropdown = driver.find_element(By.ID, "circulo")

    # Crear un objeto Select a partir del elemento "Circulo Registral"
    select = Select(dropdown)

    # Obtener todos los elementos del menú desplegable
    options = select.options

    # Crear una lista con el texto de cada opción
    options_text = [option.text for option in options]

    def select_option_by_number(number):
        global match_text
        # Buscar el elemento deseado por su número entre paréntesis
        pattern = rf'\(\s*{number}\s*\)'
        match_text = None
        for text in options_text:
            if re.search(pattern, text):
                match_text = text
                break
        if match_text is None:
            raise ValueError(f'No se encontró una opción con el número {number}')
        return match_text

    # Encontrar el elemento del menú desplegable
    dropdown = driver.find_element(By.ID, "circulo")
    # Crear un objeto Select a partir del elemento "Circulo Registral"
    select = Select(dropdown)
    # Encontrar la opción deseada por su valor y hacer clic en ella
    opcion_seleccionada = select_option_by_number(str(numero_circulo).zfill(3))
    option = driver.find_element(By.XPATH, f'//*[@id="circulo"]/option[text()="{opcion_seleccionada}"]')
    option.click()

    #Aquí se ponen los números del folio de matrícula inmobiliaria
    campo_texto = driver.find_element(By.XPATH, '//*[@id="matricula"]')
    campo_texto.send_keys(str(numero_matricula))

    # Encontrar el botón de "Buscar inmueble" por su clase
    buscar_btn = driver.find_element(By.CLASS_NAME, 'btn-danger')

    # Hacer clic en el botón
    buscar_btn.click()

    # esperar hasta que aparezcan los resultados
    wait = WebDriverWait(driver, 10)
    consultar_btn = wait.until(EC.presence_of_element_located((By.XPATH, '/html/body/div[1]/div/div[3]/div[2]/table/tbody/tr/td[1]/a')))

    # Encontrar el botón de "Consultar" por su XPath
    consultar_btn = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[3]/div[2]/table/tbody/tr/td[1]/a')

    # Hacer clic en el botón
    consultar_btn.click()

    # Esperar a que la página termine de cargar
    time.sleep(10)
```

```python
list_of_dataframes = []
for i in range(len(data)):
    orip = data['ORIP'][i] 
    folio = data['FOLIO'][i]
    print(i,folio)

    iniciar_sesion("https://www.vur.gov.co/siteminderagent/forms_es-ES/loginsnr.fcc?TYPE=100663297&REALMOID=06-6c1363d6-46ce-4693-bea4-501339aa6485&GUID=&SMAUTHREASON=0&METHOD=GET&SMAGENTNAME=-SM-fqc7JfbsitKYA98880nx7GzOWf3PHSx%2frBwpwn0hvw7giR0TRx5Ii32r0m4mIlPP&TARGET=-SM-http%3a%2f%2fwww%2evur%2egov%2eco%2fportal%2f", "", "")

    # utiliza la clase ActionChains para simular un hover sobre el submenú
    hover = ActionChains(driver).move_to_element(submenu)
    hover.perform()

    # espera unos segundos para que el submenú se despliegue
    time.sleep(2)

    # haz clic en uno de los elementos del submenú "Datos Básicos del Inmueble"
    submenu_item = driver.find_element(By.XPATH, '//*[@id="menu-navegacion"]/li[3]/ul/li/ul/li[1]/a')
    submenu_item.click()

    # espera unos segundos para que el submenú se despliegue
    time.sleep(2)

    from selenium.webdriver.support.ui import Select
    buscar_inmueble(driver, str(orip), str(folio))

    # Obtener el código fuente HTML de la página actual
    html = driver.page_source

    # Crear un objeto Beautiful Soup a partir del código fuente HTML
    soup = BeautifulSoup(html, 'html.parser')
    #print(soup)

    # Encuentra el elemento div con la clase "datos-basicos"
    datos_basicos_div = soup.find('div', {'class': 'datos-basicos'})

    # Encuentra todos los elementos div con la clase "ng-binding" dentro del elemento encontrado
    campos_div = datos_basicos_div.find_all('div', {'class': 'ng-binding'})
    #print(campos_div)

    # Crear un diccionario con las claves y valores correspondientes para cada campo de datos
    campo_datos = {
        'N° Matrícula Inmobiliaría:': 'N° Matrícula Inmobiliaría',
        'Departamento:': 'Departamento',
        'Municipio:': 'Municipio',
        'Vereda:': 'Vereda',
        'Referencia Catastral:': 'Referencia Catastral',
        'Referencia Catastral Anterior:': 'Referencia Catastral Anterior',
        'Cédula Catastral:': 'Cédula Catastral',
        'Nupre:': 'Nupre'
    }

    # Crear un diccionario vacío para almacenar los datos básicos
    datos_basicos = {}
    
    datos_basicos['ORIP'] = match_text

    # Procesa cada elemento div encontrado para extraer la información
    for campo in campos_div:
        titulo = campo.find('span', {'class': 'span-titulo'}).text.strip()
        valor = campo.text.replace(titulo, '').strip()
        if titulo in campo_datos:
            datos_basicos[campo_datos[titulo]] = valor

    ### Segunda parte de la VUR

    # Extraer la dirección actual del inmueble
    direccion_actual_html = soup.find('span', class_='span-titulo', text='Dirección Actual del Inmueble: ').parent
    datos_basicos['Dirección Actual del Inmueble'] = direccion_actual_html.text.strip().replace('Dirección Actual del Inmueble: ', '')

    # Extraer la dirección anterior del inmueble
    elementos = soup.find_all('div', class_='col-md-12')

    # Iterar por los elementos y extraer el texto de la dirección antigua
    for elemento in elementos:
        if 'Direcciones Anteriores' in elemento.text:
            direccion_ant = elemento.text.split(':')[-1].strip()
    datos_basicos['Direcciones Anteriores'] = direccion_ant

    def agregar_dato_basico(soup, texto_buscar, clave):
        # Buscar el elemento HTML que contiene la información
        elemento_html = soup.find('span', class_='span-titulo', text=texto_buscar).parent

        # Extraer el texto de ese elemento HTML
        dato = elemento_html.text.strip().replace(texto_buscar, '').replace('\n', '')

        # Reemplazar los espacios en blanco antes del guion
        dato = re.sub(r'\s(?=-)', '', dato)

        # Separar las matrículas inmobiliarias en líneas nuevas
        dato = re.sub(r'(?<=\d)(?=\d{3}-)', '\n', dato)

        # Agregar al diccionario la información como un valor asociado a la clave correspondiente
        datos_basicos[clave] = dato

    agregar_dato_basico(soup, 'Fecha de Apertura del Folio: ', 'Fecha de Apertura del Folio')
    agregar_dato_basico(soup, 'Tipo de Instrumento: ', 'Tipo de Instrumento')
    agregar_dato_basico(soup, 'Fecha de Instrumento: ', 'Fecha de Instrumento')
    agregar_dato_basico(soup, 'Estado Folio: ', 'Estado Folio')
    agregar_dato_basico(soup, 'Matrícula(s) Matriz: ', 'Matrícula(s) Matriz')
    agregar_dato_basico(soup, 'Matrícula(s) Derivada(s): ', 'Matrícula(s) Derivada(s)')
    agregar_dato_basico(soup, 'Tipo de Predio: ', 'Tipo de Predio')

    def extraer_texto(soup, texto_buscar, clave):
        # Encontrar todas las secciones que tienen la clase 'row single-child'
        secciones = soup.find_all('div', class_='row single-child')

        # Recorrer cada sección y extraer el texto
        for seccion in secciones:
            # Encontrar el título de la sección
            titulo = seccion.find('h3')

            # Comprobar si el título de la sección coincide con el texto que estamos buscando
            if titulo and texto_buscar in titulo.text:
                # Encontrar el texto dentro de la sección
                texto = seccion.find('div', class_='col-lg-11 col-md-11 pull-right ng-binding').text.strip()

                # Guardar el texto en el diccionario usando el título como clave
                datos_basicos[clave] = texto

    extraer_texto(soup, 'Complementaciones', 'Complementaciones')
    extraer_texto(soup, 'Cabidad y Linderos', 'Cabidad y Linderos')

    #Almacena la información en dataframes
    df1 = pd.DataFrame(datos_basicos, index=[0])
    df1['Dirección Actual del Inmueble'] = df1['Dirección Actual del Inmueble'].str.lstrip()
    df1['Fecha de Apertura del Folio'] = df1['Fecha de Apertura del Folio'].str.lstrip()
    df1['Tipo de Instrumento'] = df1['Tipo de Instrumento'].str.lstrip()
    df1['Fecha de Instrumento'] = df1['Fecha de Instrumento'].str.lstrip()
    df1['Estado Folio'] = df1['Estado Folio'].str.lstrip()
    #df1

    # cambiar de nuevo al contexto predeterminado fuera del iframe
    driver.switch_to.default_content()

    # encuentra el enlace "Salir" y haz clic en él
    salir_link = driver.find_element(By.XPATH, '//*[@id="menu-navegacion"]/li[5]/a')
    salir_link.click()

    iniciar_sesion("https://www.vur.gov.co/siteminderagent/forms_es-ES/loginsnr.fcc?TYPE=100663297&REALMOID=06-6c1363d6-46ce-4693-bea4-501339aa6485&GUID=&SMAUTHREASON=0&METHOD=GET&SMAGENTNAME=-SM-fqc7JfbsitKYA98880nx7GzOWf3PHSx%2frBwpwn0hvw7giR0TRx5Ii32r0m4mIlPP&TARGET=-SM-http%3a%2f%2fwww%2evur%2egov%2eco%2fportal%2f", "julio.fonseca", "Julio07*")

    # utiliza la clase ActionChains para simular un hover sobre el submenú
    hover = ActionChains(driver).move_to_element(submenu)
    hover.perform()

    # espera unos segundos para que el submenú se despliegue
    time.sleep(2)

    # haz clic en uno de los elementos del submenú "Estado Jurídico del Inmueble"
    submenu_item = driver.find_element(By.XPATH, '//*[@id="menu-navegacion"]/li[3]/ul/li/ul/li[2]/a')
    submenu_item.click()

    # espera unos segundos para que el submenú se despliegue
    time.sleep(2)

    from selenium.webdriver.support.ui import Select
    buscar_inmueble(driver, str(orip), str(folio))

    # encuentra el enlace "Lista" y haz clic en él
    lista = driver.find_element(By.XPATH, '/html/body/div[1]/div/div[11]/div/div[3]/ul/li[2]/a')
    lista.click()

    # Obtener el código fuente HTML de la página actual
    html = driver.page_source

    # Crear un objeto Beautiful Soup a partir del código fuente HTML
    soup = BeautifulSoup(html, 'html.parser')

    # Encontrar todos los elementos div con la clase "col-md-4 ng-binding"
    elementos = soup.find_all('div', class_='col-md-4 ng-binding')

    # Iterar por los elementos y extraer el texto de la matrícula inmobiliaria
    for elemento in elementos:
        if 'No. Matricula Inmobiliaria' in elemento.text:
            matricula = elemento.text.split(':')[-1].strip()

    # encontrar todas las divs con la clase 'anotacion'
    anotaciones = soup.find_all('div', class_='anotacion')

    patron_nro = r'ANOTACION: ([\w\s\d-]+) Fecha'
    patron_fecha = r'Fecha: ([\d-]+)'
    patron_radicacion = r'Radicación: ([\w\d-]+)'
    patron_doc = r'Doc: (.+?)VALOR ACTO'
    patron_valor_acto = r'VALOR ACTO: (\S+)ESPECIFICACION:'
    patron_especificacion = r'ESPECIFICACION: (.+?)PERSONAS'
    pattern1 = r'DE: (.+)'
    pattern2 = r'A: (.+)'

    # inicializar una lista para guardar los datos de cada anotación
    datos = []

    for anotacion in anotaciones:
        resultado = {}
        resultado['FMI'] = matricula
        # Buscar los patrones y almacenar los valores en un diccionario
        anotacion_match = re.search(patron_nro, str(anotacion))
        if anotacion_match:
            resultado['Anotación'] = anotacion_match.group(1)
        else:
            resultado['Anotación'] = None
        fecha_match = re.search(patron_fecha, str(anotacion))
        if fecha_match:
            resultado['Fecha'] = fecha_match.group(1)
        else:
            resultado['Fecha'] = None
        radicacion_match = re.search(patron_radicacion, str(anotacion))
        if radicacion_match:
            resultado['Radicacion'] = radicacion_match.group(1)
        else:
            resultado['Radicacion'] = None
        documento_match = re.search(patron_doc, str(anotacion))
        if documento_match:
            resultado['Documento'] = documento_match.group(1)
        else:
            resultado['Documento'] = None
        valor_acto_match = re.search(patron_valor_acto, str(anotacion))
        if valor_acto_match:
            resultado['Valor Acto'] = valor_acto_match.group(1)
        else:
            resultado['Valor Acto'] = None
        especificacion_match = re.search(patron_especificacion, str(anotacion))
        if especificacion_match:
            resultado['Especificación'] = especificacion_match.group(1)
        else:
            resultado['Especificación'] = None 

        # buscar los nombres de las partes
        de_nombres = []
        for line in anotacion.find_all(text=re.compile('DE:')):
            match = re.search(pattern1, line)
            if match:
                nombre_completo = match.group(1)
                de_nombres.append(nombre_completo)

        a_nombres = []
        for line in anotacion.find_all(text=re.compile('A:')):
            match = re.search(pattern2, line)
            if match:
                nombre_completo = match.group(1)
                a_nombres.append(nombre_completo)    

        # agregar los nombres de las partes al diccionario resultado
        resultado['De Nombres'] = de_nombres
        resultado['A Nombres'] = a_nombres

        # agregar los datos de la anotación a la lista general
        datos.append(resultado)

    # cambiar de nuevo al contexto predeterminado fuera del iframe
    driver.switch_to.default_content()

    # encuentra el enlace "Salir" y haz clic en él
    salir_link = driver.find_element(By.XPATH, '//*[@id="menu-navegacion"]/li[5]/a')
    salir_link.click()

    # crear un DataFrame de pandas a partir de la lista de datos
    df2 = pd.DataFrame(datos)
    df = pd.concat([df1,df2],axis=1)
    df = df.fillna(method='ffill')
    list_of_dataframes.append(df)
combined_df = pd.concat(list_of_dataframes) 
```
